```python
import pandas as pd

df = pd.read_csv(
    '/content/twitter_training.csv',
    header=None,
    names=['ID', 'Entity', 'Sentiment', 'Tweet'],
    on_bad_lines='skip',
    engine='python'
)

print(df.head())
print(df.shape)
```

```python
print(df['Sentiment'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Sentiment'].value_counts().plot(kind='bar')

plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')

plt.show()
```

```python
df['Sentiment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Sentiment Analysis')
plt.ylabel('')

plt.show()
```

```python
top_entities = df['Entity'].value_counts().head(10)

top_entities.plot(kind='bar')

plt.title('Top 10 Mentioned Entities')
plt.xlabel('Entity')
plt.ylabel('Count')

plt.show()
```
