```python
import pandas as pd

df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_EN_csv_v2_327505.csv', skiprows=4)
print(df.head())
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

# Remove rows where 2023 population is missing
df_clean = df[['Country Name', '2023']].dropna()

# Get top 10 countries by population
top10 = df_clean.sort_values(by='2023', ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2023'])
plt.title('Top 10 Most Populated Countries (2023)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
